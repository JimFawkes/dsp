# Chapter 4, Ex 2

The number generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from random.random and plot their `PMF` and `CDF`. Is the distribution uniform?

```python
import random
NUMBER_COUNT=1000
ROUND_TO_POSITION=1
```

```python
# Generate a list with 1000 random variables
ml = [random.random() for x in range(NUMBER_COUNT)]
```

```python
# Instantiate the Hist Distribution
random_hist = Hist(ml)
```

```python
random_hist.cdf.percentile(50)
```

```python
mean = random_hist.mean
mean
```

```python
quants = random_hist.cdf.quantile(10)
quants
```

# MISSING PLOTS

```python
# Verify that for the quants the step and percentile are similar.
for key, value in quants:
    assert round(key, ROUND_TO_POSITION) == round(value, ROUND_TO_POSITION), "Quants Test Failed: The distribution does not seem to be uniformly distributed."

# Verify that the mean is close to 0.5
assert round(mean, ROUND_TO_POSITION) == 0.5, "Mean Test Failed: The distribution does not seem to be uniformly distributed."
```
